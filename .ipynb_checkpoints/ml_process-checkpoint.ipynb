```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats

from sklearn.preprocessing import MinMaxScaler
```

```python
train_df = pd.read_csv('delivery_ml.csv')
train_df.head()
```

```python
train_df = train_df.drop(columns=['Unnamed: 0'])

train_df
```

```python
train_df.isna().sum()
```

```python
feature_list = ['amount', 'weight', 'delivery_time']

for col in feature_list:
    
    plt.figure(figsize=(16, 4))
    
    plt.subplot(1, 3, 1)
    sns.histplot(train_df[col], bins = 30)
    plt.title('Histogram')
    
    
    plt.show()
    
```

```python
sns.distplot(train_df['amount'], kde = False, color ='red')
plt.show()
```

```python
#Transform it into log-scale

train_df['amount'] = np.log(train_df['amount'])
```

```python
#Lihat grafik amount yang baru

sns.distplot(train_df['amount'], kde = False, color ='red')
plt.show()
```

```python
train_df.drop(['latitude','longitude'],1,inplace=True)
```

```python
train_df.drop(['date_stop'],1,inplace=True)
```

```python
train_df.drop(['cod_received_status', 'task_status_label', 'task_detail_status_label'],1,inplace=True)
```

```python
train_df.drop(['flow', 'task_status'],1,inplace=True)
```

```python
train_df
```

```python
train_df.columns
```

```python
#Feature Encoding
!pip install feature-engine
```

```python
from feature_engine.encoding import OrdinalEncoder
from feature_engine.encoding import OneHotEncoder
```

```python
X = train_df.drop(['amount'], axis = 1)
y = train_df['amount']
```

```python
X
```

```python
ordinal_enc = OrdinalEncoder(
    encoding_method='arbitrary',
    variables=['branch_origin', 'branch_destination']
)
```

```python
X[['branch_origin', 'branch_destination']]
```

```python
ordinal_enc.fit(X)

X = ordinal_enc.transform(X)
```

```python
X[['branch_origin', 'branch_destination']]
```

```python
X.drop(['task_status_id','task_assigned_to', 'task_id'],1,inplace=True)

X
```

```python
X.drop(['task_detail_status'],1,inplace=True)
X
```

```python
X.drop(['date_start', 'time_start', 'time_stop'],1,inplace=True)

X
```

```python
#With receiver city
contoh = X.copy()

ohe_enc = OneHotEncoder(top_categories=None) 
ohe_enc.fit(contoh)
contoh = ohe_enc.transform(X)

contoh
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(contoh, y, test_size=0.2, random_state = 99)
```

```python
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

```python
y_train
```

```python
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn.ensemble import AdaBoostRegressor

from sklearn.ensemble import HistGradientBoostingRegressor

from sklearn.metrics import mean_squared_error

gbr = GradientBoostingRegressor(n_estimators=500, 
                                learning_rate=0.1, 
                                max_depth=8, 
                                random_state=99,
                                loss='squared_error', 
                                min_samples_split=5
                               )

frs = RandomForestRegressor(n_estimators=500,
                                max_depth=8, 
                                random_state=99,
                                criterion='squared_error', 
                                min_samples_split=5)

ada = AdaBoostRegressor(n_estimators=500,
                                learning_rate=1.0,
                                random_state=99,
                                loss='linear')

hst = HistGradientBoostingRegressor(loss='squared_error',
                                   learning_rate=0.1,
                                   max_depth=8)

```

```python
gbr.fit(X_train, y_train)
```

```python
frs.fit(X_train, y_train)
```

```python
ada.fit(X_train, y_train)
```

```python
hst.fit(X_train, y_train)
```

```python
print(gbr.score(X_train, y_train))
print(gbr.score(X_valid, y_valid))
```

```python
print(frs.score(X_train, y_train))
print(frs.score(X_valid, y_valid))
```

```python
print(ada.score(X_train, y_train))
print(ada.score(X_valid, y_valid))
```

```python
print(hst.score(X_train, y_train))
print(hst.score(X_valid, y_valid))
```
