```python
groupped_data = date_03.groupby(['task_detail_status_label']).agg({'amount': 'sum'})
groupped_data["%"] = groupped_data.apply(lambda x:  100*x / x.sum())
groupped_data
```

```python
#Visualize Data
ax = groupped_data.plot(
    y='%', kind='bar',
    title='Loss Ammount of COD based on several factors',
    ylim=[0, 30],
    label='Percent'
)

# Add Labels on Top of Bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%')

plt.show()
```
